# Indian Urban AQI and Health Impact Analysis (2019–2024)

This notebook explores the AQI levels and health impact scores across major Indian cities based on the dataset provided.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load dataset
df = pd.read_csv("/kaggle/input/indian-urban-aqi-and-health-impact-cpcb-20192024/indian_aqi_health_impact_2019_2024.csv")
df.head()
```

```
# AQI Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["AQI"], bins=30, kde=True)
plt.title("Distribution of AQI Across Indian Cities (2019–2024)")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```
# AQI vs Health Impact Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="AQI", y="Health Impact Score", hue="City", alpha=0.7)
plt.title("AQI vs Health Impact Score")
plt.xlabel("Air Quality Index (AQI)")
plt.ylabel("Health Impact Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
